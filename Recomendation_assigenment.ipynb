#creating pivot table

```python
book1.sort_values('user_id',inplace=True)
book1
```

```python
user_book=book1.pivot_table(index='book',columns='user_id',values='rating').reset_index(drop=True)
user_book
```

```python
user_book.fillna(0,inplace=True)
user_book
```

#calculating the similarities

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
book_sim=1-pairwise_distances(user_book.values,metric='cosine')
book_sim
```

#sorting the values into dataframe

```python
book_sim_1=pd.DataFrame(book_sim)
book_sim_1
```

```python
book_sim_1.index=book1.book.unique()
book_sim_1.columns=book1.book.unique()
```

```python
book_sim_1
```

```python
book_sim_1.iloc[0:7,0:7]
```

#filling diagonal values with 0

```python
np.fill_diagonal(book_sim, 0)
```

```python
book_sim_1
```
